```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
# Loading dataset
df = pd.read_csv("/Users/admin/Downloads/amz_uk_price_prediction_dataset.csv")
df
```

```python
# Create crosstab between 'category' and 'isBestSeller'
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
# Calculate the proportion of best-sellers (True = 1) in each category
crosstab_result['proportionBestSeller'] = crosstab_result[True] / crosstab_result.sum(axis=1)

# Sort categories based on the proportion of best-sellers in descending order
sorted_categories = crosstab_result.sort_values(by='proportionBestSeller', ascending=False)

# Display the sorted DataFrame
print(sorted_categories[['proportionBestSeller']])
```

```python
from scipy.stats import chi2_contingency
# Prepare the contingency table
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(crosstab_result)
# Output the Chi-square test result
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
#print(f"Degrees of Freedom: {dof}")
```

```python
# Chi-square Test: Helps determine if there’s an association between the category and best-seller status.
#Cramér's V: Provides the strength of this association.
#By conducting these tests, one gains statistical insights into how categories and best-seller status are related in the dataset. 
```

```python
# Total number of observations
number_observations = crosstab_result.values.sum()
# Calculate Cramér's V
cramers_v = np.sqrt(chi2 / (number_observations * (min(crosstab_result.shape) - 1)))
print(f"Cramér's V: {cramers_v}")
```
